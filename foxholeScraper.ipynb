```python
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup
import os
import pandas as pd
import re
```

```python
# https://medium.com/@tgcajal/beautiful-soup-template-and-example-for-python-web-scraping-355ccfe9eca
# target = https://foxhole.fandom.com/wiki/Vehicle_Health

url = 'https://foxhole.fandom.com/wiki/Vehicle_Health'
req = Request(url=url,headers={'user-agent': 'my-app/0.0.1'})
response = urlopen(req)
html = BeautifulSoup(response)
```

```python
names = []
for idx, table_title in enumerate(html.find_all('td')):
    try:
        name = table_title.find('a').find('img').get('data-image-name')
        name = name.replace(" Vehicle Icon.png","")
        names.append(name)
    except:
        continue
```
